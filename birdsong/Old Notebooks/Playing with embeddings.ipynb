```python
import os
import numpy as np

path = 'BirdCLEF/birdnet_embeddings'

embeddings = {}

for f in sorted(os.listdir(path))[:10]:
    species_id = f.split('-')[0]
    
    emb = np.load('%s/%s' % (path, f))
    
    if species_id in embeddings:
        np.append(embeddings[species_id], emb)
    else:
        embeddings[species_id] = emb
```

```python
from IPython.display import clear_output
import librosa

path = 'BirdCLEF/train_audio'

lengths = {}

i = 0
for folder in sorted(os.listdir(path)):
    total_length = 0
    
    for file in os.listdir('%s/%s' % (path, folder)):
        filename = '%s/%s/%s' % (path, folder, file)
        
        clear_output(wait = True)
        print('Processing file %d/%d' % (i, 16941))
        i += 1
        
        sig, _ = librosa.load(filename)
        
        total_length += len(sig)
        
    lengths[folder] = total_length
```
